```python
from easy1 import encrypt, mux, demux, sbox, pbox
```

```python
def f_spn(p):
    u = []
    # Run through the s-boxes
    for x in demux(p):
        u.append(sbox(x))
    # Run through the p-box and return
    return pbox(mux(u))
```

```python
# Read the file with PT-CT pairs
PC = []
pairs_file = "plain-cipher.txt"
f = open(pairs_file, "r")
line = f.readline()
while line != '':
    pt, ct = line.strip().split(sep=" ", maxsplit=1)
    PC.append((int(pt), int(ct)))
    line = f.readline()
f.close()
```

```python
# PC = PC[:100]
```

```python
for inx, pair in enumerate(PC):
    print('Running pair# %d out of %d pairs' % (inx+1, len(PC)))
    p, c = pair
    fp, fc = f_spn(p), f_spn(c)
    PC_copy = PC.copy()
    PC_copy.remove(pair)
    for q, d in PC_copy:
        if fp ^ q == fc ^ d:
            print('Checking for collision')
            k = (fp ^ q) & 0xfffffffff
            print("Key:", k)
            if encrypt(p, 20, k) == c:
                print('Found key!')
                break
            else:
                print('False positive!')
```
