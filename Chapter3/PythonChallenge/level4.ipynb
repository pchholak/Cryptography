```python
def get_page_source_object(url):
    html_object = urllib.request.urlopen(url)
    return html_object

def get_text_from_url(url):
    html_object = get_page_source_object(url)
    raw = BeautifulSoup(html_object, 'html.parser').get_text()
    return raw
```

```python
get_text_from_url(home_url)
```

```python
nothing, current_url, text = [], [], []
current_url.append(home_url)
print("0")
html_object = get_page_source_object(current_url[0])
soup = BeautifulSoup(html_object)
link = soup.findAll('a')[0]
nothing.append(link.get('href')[-5:])
print("nothing = ", nothing[0])
current_url.append(home_url + "?" + "nothing=" + nothing[0])
print(current_url[1])
i, run = 0, True
while (i<400) and run:
    i += 1
    print(i)
    raw = get_text_from_url(current_url[i])
    text.append(raw)
    number = word_tokenize(raw)[-1]
    if re.search("^[0-9]+$", number):
        nothing.append(number)
        print("nothing = ", number)
        current_url.append(home_url + "?" + "nothing=" + number)
        print(current_url[-1])
    else:
        run = False
        print("For i -> {}, Text -> {}".format(i, text[-1]))
# next_url = current_url[-1]
```
