```python
####################################################
# I/O helper functions
####################################################
string2hex = lambda x: "".join((hex(ord(c))[2:].zfill(2) for c in x))

def text2state(text):
    '''
    This function takes a 128-bit block and translates it into
    a state matrix where each column is a 4-byte word
    The inputs should be hex presented as strings
    '''
    state = []
    for i in range(16):
        byte = int(text[i * 2 : (i * 2) + 2], 16) #Operates on two hex chars at a time
        if i % 4 == 0:
            state.append([byte])
        else:
            state[i // 4].append(byte)
    return state

def state2text(state):
    '''
    Reverses text2state to output either ciphertext or plaintext as a
    HEX string (ex: '5233ffaa')
    '''
    text = ''
    for column in state:
        for byte in column:
            val1 = "%x" % (byte >> 4)
            val2 = "%x" % (byte & 0xF)
            text = text + val1 + val2
    return text

####################################################
# Key schedule functions
####################################################

def rot_words(w):
    w[0], w[1], w[2], w[3] = w[1], w[2], w[3], w[0]
    return w

def sub_words(w):
    sbox = AES_tables.SBOX.value
    for i in range(4):
        w[i] = sbox[w[i]]
    return w

def _key_expansion(key, Nk=4):
    Nb = 4
    Nr = 10
    w = text2state(key)
    i = Nk
    while len(w) < Nb * (Nr + 1):
        w.append([0, 0, 0, 0])
    
    rcon = AES_tables.RCON.value
    while i < Nb * (Nr + 1):
        temp = w[i - 1][:]
        if i % Nk == 0:
            temp = sub_words(rot_words(temp))
            temp[0] = temp[0] ^ rcon[i // Nk]
        for x in range(4):
            w[i][x] = w[i - Nk][x] ^ temp[x]
        i += 1
    return w

####################################################
# Round functions
####################################################

def _sub_bytes(S):
    sbox = AES_tables.SBOX.value
    for i in range(4):
        for j in range(4):
            S[i][j] = sbox[S[i][j]]
    return S

def _inv_sub_bytes(S):
    sbox = AES_tables.SBOX_INV.value
    for i in range(4):
        for j in range(4):
            S[i][j] = sbox[S[i][j]]
    return S

def _shift_rows(s):
    s[0][1], s[1][1], s[2][1], s[3][1] = s[1][1], s[2][1], s[3][1], s[0][1]
    s[0][2], s[1][2], s[2][2], s[3][2] = s[2][2], s[3][2], s[0][2], s[1][2]
    s[0][3], s[1][3], s[2][3], s[3][3] = s[3][3], s[0][3], s[1][3], s[2][3]
    return s

def _inv_shift_rows(s):
    s[0][1], s[1][1], s[2][1], s[3][1] = s[3][1], s[0][1], s[1][1], s[2][1]
    s[0][2], s[1][2], s[2][2], s[3][2] = s[2][2], s[3][2], s[0][2], s[1][2]
    s[0][3], s[1][3], s[2][3], s[3][3] = s[1][3], s[2][3], s[3][3], s[0][3]
    return s

def xtime(a, n=1):
    for i in range(n):
        if a & 0x80:
            a = a << 1
            a ^= 0x1B
        else:
            a = a << 1
    return a & 0xFF

def mix_a_column(s):
    u = s[0] ^ s[1] ^ s[2] ^ s[3]
    v = s[0]
    # XORing s[i] to itself cancels out the extra term in 'u'
    s[0] ^= xtime(s[0] ^ s[1]) ^ u
    s[1] ^= xtime(s[1] ^ s[2]) ^ u
    s[2] ^= xtime(s[2] ^ s[3]) ^ u
    s[3] ^= xtime(s[3] ^ v) ^ u
    return s

def inv_mix_a_column(s):
    x = list(s)
    u = xtime(s[0], 3) ^ xtime(s[1], 3) ^ xtime(s[2], 3) ^ xtime(s[3], 3)
    x[0] = s[3] ^ xtime(s[0], 2) ^ xtime(s[0]) ^ xtime(s[1]) ^ s[1] ^ xtime(s[2], 2) ^ s[2] ^ u
    x[1] = s[0] ^ xtime(s[1], 2) ^ xtime(s[1]) ^ xtime(s[2]) ^ s[2] ^ xtime(s[3], 2) ^ s[3] ^ u
    x[2] = s[1] ^ xtime(s[2], 2) ^ xtime(s[2]) ^ xtime(s[3]) ^ s[3] ^ xtime(s[0], 2) ^ s[0] ^ u
    x[3] = s[2] ^ xtime(s[3], 2) ^ xtime(s[3]) ^ xtime(s[0]) ^ s[0] ^ xtime(s[1], 2) ^ s[1] ^ u
    return x

def _mix_columns(s):
    for i in range(4):
        s[i] = mix_a_column(s[i])
    return s

def _inv_mix_columns(s):
    for i in range(4):
        s[i] = inv_mix_a_column(s[i])
    return s

def _add_round_key(state, k):
    for i in range(4):
        for j in range(4):
            state[i][j] ^= k[i][j]
    return state

def encrypt(text, key):
    Nb, Nr = 4, 10

    S = text2state(text)
    print("Input state:", S)
    w = _key_expansion(key)
    S = _add_round_key(S, w[0:Nb])
    print("After first key addition:", S)

    for i in range(1, 10):
        S = _sub_bytes(S)
        S = _shift_rows(S)
        S = _mix_columns(S)
        S = _add_round_key(S, w[i * Nb : (i + 1) * Nb])

    S = _sub_bytes(S)
    S = _shift_rows(S)
    S = _add_round_key(S, w[Nr * Nb : (Nr + 1) * Nb])

    return state2text(S)

def decrypt(text, key):
    Nb, Nr = 4, 10
    C = text2state(text)
    z = _key_expansion(key)
    C = _add_round_key(C, z[40:])
    
    for i in range(9, 0, -1):
        C = _inv_shift_rows(C)
        C = _inv_sub_bytes(C)
        C = _add_round_key(C, z[i * Nb : (i + 1) * Nb])
        C = _inv_mix_columns(C)
    
    C = _inv_shift_rows(C)
    C = _inv_sub_bytes(C)
    C = _add_round_key(C, z[0:Nb])
    
    return state2text(C)
```
