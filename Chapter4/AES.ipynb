```python
text = b'a secret message'
def text2state(text):
    num = int.from_bytes(text, byteorder="little")
    state = []
    for i in range(16):
        byte = (num >> (8 * (15 - i))) & 0xFF
        if i % 4 == 0:
            state.append([byte])
        else:
            state[i // 4].append(byte)
    return state
S = text2state(text)
print(S)
```

```python
def _sub_bytes(S0):
    sbox = AES_tables.SBOX.value
    S1 = S0
    for i in range(4):
        for j in range(4):
            S1[i][j] = sbox[S0[i][j]]
    return S1
S = _sub_bytes(S)
print(S)
```

```python
def _shift_rows(s):
    s[0][1], s[1][1], s[2][1], s[3][1] = s[1][1], s[2][1], s[3][1], s[0][1]
    s[0][2], s[1][2], s[2][2], s[3][2] = s[2][2], s[3][2], s[0][2], s[1][2]
    s[0][3], s[1][3], s[2][3], s[3][3] = s[3][3], s[0][3], s[1][3], s[2][3]
    return s
S = _shift_rows(S)
print(S)
```
