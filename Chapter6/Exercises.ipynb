```python
def generate_linear_expressions(M, n):
    '''
    Generates sorted linear expressions with higher bias
    '''
    N = 2 ** n - 1
    V = M.ravel()
    V = np.abs(V)
    ids = (-V).argsort()
    expr = []
    for i in ids:
        r = i // N
        c = i % N
        x, y = int2bin(r, n), int2bin(c, n)
        expr.append((x, y, M[r, c]))
    return expr

def bits2exp(t, var):
    '''
    Converts bitstring tuple 't' to linear expressions of variable 'var'
    '''
    s = ''
    for i, b in enumerate(t):
        s += (var + str(i) + ' + ') * b
    if s == '':
        return '0'
    else:
        return s[:-3]
```
